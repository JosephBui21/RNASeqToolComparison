```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''converts the string version of H:MM:SS string into seconds'''
def get_seconds(str_time):
    str_time = str_time.strip('\n')
    h, m, s = str_time.split(':')
    return (int(h) * 3600) + (int(m) * 60) + float(s) 
```

```python
#gets the averages of the timings from the 10 versions in each sample number & synthetic dataset
def get_avg_timings(timings):
    #timings for 2 samples per cond
    count_2 = 0
    #timings for 5 samples per cond
    count_5 = 0
    #timings for 10 samples per cond
    count_10 = 0
    for tup in timings:
        num_samp = tup[0]
        val = tup[1]
        if num_samp == '2':
            count_2 += val
        elif num_samp == '5':
            count_5 += val
        elif num_samp == '10':
            count_10 += val
    avg_2 = round(count_2/10, 2)
    avg_5 = round(count_5/10, 2)
    avg_10 = round(count_10/10, 2)
    return [('2', avg_2), ('5', avg_5), ('10', avg_10)]
```

```python
#open the log.txt file
file = open('log.txt', 'r')
lines = file.readlines()

#Function to gather timings for 11 synthetic datasets for all samples, so each list below will have
#33 (10 versions each sample) tuples signifying (num samples, time to complete tool on the synthetic data)
def get_timings(tool):
    baseline0_0 = []
    baseline1250_0 = []
    baseline625_625 = []
    baseline4000_0 = []
    baseline2000_2000 = []
    poisson0_0 = []
    poisson625_625 = []
    single0_0 = []
    single625_625 = []
    random0_0 = []
    random625_625 = []
    tool_timings_all = None

    for line in lines:
        time_str = line[-15:]
        if tool in line:
            if "baseline0_0_2" in line:
                baseline0_0.append(('2', get_seconds(time_str)))
            if "baseline0_0_5" in line:
                baseline0_0.append(('5', get_seconds(time_str)))
            if "baseline0_0_10" in line:
                baseline0_0.append(('10', get_seconds(time_str)))
            if "baseline1250_0_2" in line:
                baseline1250_0.append(('2', get_seconds(time_str)))
            if "baseline1250_0_5" in line:
                baseline1250_0.append(('5', get_seconds(time_str)))
            if "baseline1250_0_10" in line:
                baseline1250_0.append(('10', get_seconds(time_str)))
            if "baseline625_625_2" in line:
                baseline625_625.append(('2', get_seconds(time_str)))
            if "baseline625_625_5" in line:
                baseline625_625.append(('5', get_seconds(time_str)))
            if "baseline625_625_10" in line:
                baseline625_625.append(('10', get_seconds(time_str)))
            if "baseline4000_0_2" in line:
                baseline4000_0.append(('2', get_seconds(time_str)))
            if "baseline4000_0_5" in line:
                baseline4000_0.append(('5', get_seconds(time_str)))
            if "baseline4000_0_10" in line:
                baseline4000_0.append(('10', get_seconds(time_str)))
            if "baseline2000_2000_2" in line:
                baseline2000_2000.append(('2', get_seconds(time_str)))
            if "baseline2000_2000_5" in line:
                baseline2000_2000.append(('5', get_seconds(time_str)))
            if "baseline2000_2000_10" in line:
                baseline2000_2000.append(('10', get_seconds(time_str)))
            if "poisson0_0_2" in line:
                poisson0_0.append(('2', get_seconds(time_str)))
            if "poisson0_0_5" in line:
                poisson0_0.append(('5', get_seconds(time_str)))
            if "poisson0_0_10" in line:
                poisson0_0.append(('10', get_seconds(time_str)))
            if "poisson625_625_2" in line:
                poisson625_625.append(('2', get_seconds(time_str)))
            if "poisson625_625_5" in line:
                poisson625_625.append(('5', get_seconds(time_str)))
            if "poisson625_625_10" in line:
                poisson625_625.append(('10', get_seconds(time_str)))
            if "single0_0_2" in line:
                single0_0.append(('2', get_seconds(time_str)))
            if "single0_0_5" in line:
                single0_0.append(('5', get_seconds(time_str)))
            if "single0_0_10" in line:
                single0_0.append(('10', get_seconds(time_str)))
            if "single625_625_2" in line:
                single625_625.append(('2', get_seconds(time_str)))
            if "single625_625_5" in line:
                single625_625.append(('5', get_seconds(time_str)))
            if "single625_625_10" in line:
                single625_625.append(('10', get_seconds(time_str)))
            if "random0_0_2" in line:
                random0_0.append(('2', get_seconds(time_str)))
            if "random0_0_5" in line:
                random0_0.append(('5', get_seconds(time_str)))
            if "random0_0_10" in line:
                random0_0.append(('10', get_seconds(time_str)))
            if "random625_625_2" in line:
                random625_625.append(('2', get_seconds(time_str)))
            if "random625_625_5" in line:
                random625_625.append(('5', get_seconds(time_str)))
            if "random625_625_10" in line:
                random625_625.append(('10', get_seconds(time_str)))
                
    #get the average of the timings for the 10 versions for baseline0_0 for all sample numbers
    avg_baseline0_0 = get_avg_timings(baseline0_0)
    
    #get the average of the timings for the 10 versions for baseline1250_0 for all sample numbers
    avg_baseline1250_0 = get_avg_timings(baseline1250_0)
    
    #get the average of the timings for the 10 versions for baseline625_625 for all sample numbers
    avg_baseline625_625 = get_avg_timings(baseline625_625)
    
    #get the average of the timings for the 10 versions for baseline4000_0 for all sample numbers
    avg_baseline4000_0 = get_avg_timings(baseline4000_0)
    
    #get the average of the timings for the 10 versions for baseline2000_2000 for all sample numbers
    avg_baseline2000_2000 = get_avg_timings(baseline2000_2000)
    
    #get the average of the timings for the 10 versions for poisson0_0 for all sample numbers
    avg_poisson0_0 = get_avg_timings(poisson0_0)
    
    #get the average of the timings for the 10 versions for poisson625_625 for all sample numbers
    avg_poisson625_625 = get_avg_timings(poisson625_625)
    
    #get the average of the timings for the 10 versions for single0_0 for all sample numbers
    avg_single0_0 = get_avg_timings(single0_0)
    
    #get the average of the timings for the 10 versions for single625_625 for all sample numbers
    avg_single625_625 = get_avg_timings(single625_625)
    
    #get the average of the timings for the 10 versions for random0_0 for all sample numbers
    avg_random0_0 = get_avg_timings(random0_0)
    
    #get the average of the timings for the 10 versions for random625_625 for all sample numbers
    avg_random625_625 = get_avg_timings(random625_625)
    
    tool_timings_all = [avg_baseline0_0, avg_baseline1250_0, avg_baseline625_625, avg_baseline4000_0,
                       avg_baseline2000_2000, avg_poisson0_0, avg_poisson625_625, avg_single0_0,
                       avg_single625_625, avg_random0_0, avg_random625_625]
    return tool_timings_all
```

```python
#function to extract timings for desired number of samples per condition
def extract(array, num):
    res = []
    for l in array:
        for tup in l:
            if tup[0] == num:
                res.append(tup[1])
    return res
```

```python
#create bar graph for comparing tool timings for 2 samples per condition
DESeq2_2samp = extract(DESeq2, '2')
edgeR_2samp = extract(edgeR, '2')
voom_2samp = extract(voom_limma, '2')
NOISeq_2samp = extract(NOISeq, '2')
ttest_2samp = extract(ttest, '2')
PoissonSeq_2samp = extract(PoissonSeq, '2')
ABSSeq_2samp = extract(ABSSeq, '2')

index = ['baseline0_0', 'baseline1250_0', 'baseline625_625', 'baseline4000_0', 'baseline2000_2000', 'poisson0_0', 
        'poisson625_625', 'single0_0', 'single625_625', 'random0_0', 'random625_625']
df_2samp = pd.DataFrame({'DESeq2': DESeq2_2samp,
                  'NOISeq': NOISeq_2samp,
                  'edgeR': edgeR_2samp,
                  'ttest': ttest_2samp,
                  'voom.limma': voom_2samp,
                         'PoissonSeq': PoissonSeq_2samp,
                        'ABSSeq': ABSSeq_2samp}, index = index)
df_2samp

ax_2samp = df_2samp.plot.bar(rot=45, figsize = (6, 4),
                 ylabel = 'Duration (in seconds)', xlabel = 'Synthetic Dataset', 
                 title = 'Comparing Duration of Running Tools on Synthetic Datasets for 2 samples per condition')
```
